```python
# Excercise 1

import pandas as pd
data = pd.read_csv("/Users/patrickrs/Documents/GitLab/patrick-steiner/01-DS-Toolkit/Day-3-Numpy-Pandas/livecoding/Pandas Exercises/ign.csv", )

data.dtypes
```

```python
data['release_date'] = pd.to_datetime(data['release_year']*10000 + data['release_month']*100 + data['release_day'], format='%Y%m%d')
```

```python
# What’s the first game ever released?
import datetime
min_date = data['release_date'].min() 
data[data['release_date'] == min_date]

# 1970 looks wrong. Let's see the other values. One may use sort as well:
data.sort_values('release_date').head()
```

```python
# Do game scores vary from year to year?
grouped_data = pd.DataFrame(data.groupby(['release_year']).mean())
grouped_data = grouped_data['score']
grouped_data
grouped_data.plot()
```

```python
# What’s the platform with the highest average reviews?
platforms = data.groupby(['platform']).mean()
platforms = platforms['score']
max_score = platforms.max()
platforms[platforms == max_score]
```

```python
# Are blockbuster games released at a specific time of the year?
block = data.groupby(['release_month']).mean()
block = block['score']
block.plot()
```

```python
# Bonus Excercise
sales = pd.read_csv("vgsales.csv")
sales
```

```python
sales.describe()
```

```python
# Checking for missing values
sales.isnull().values.any()
# Should clean it, but let's keep going.
```

```python
# Who sold the most:
sales.groupby(["Publisher"]).sum().sort_values("Global_Sales", ascending=False)
```

```python
# Which region sold the most:
group = sales[["NA_Sales", "EU_Sales", "JP_Sales", "Global_Sales"]].sum()
group
```

```python
group.plot.pie(label = "Sales per Region", figsize = (6, 6))
```

```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook as tqdm # this is a fancy progress bar!
from time import sleep
import numpy as np
from datetime import datetime
import pandas as pd
```

```python
link = "https://www.atpworldtour.com/en/rankings/singles"
```

```python
response = requests.get(link, timeout = 15)
```

```python
response.status_code # 200 is good anything with 4 or 5 is bad. 
#check http status code when you use request sometimes fails https://restfulapi.net/http-status-codes/
```

```python
bs=BeautifulSoup(response.content, features = "html")
```

```python
dates = bs.findAll('ul', class_ = "dropdown")
dates = dates[0]
```

```python
dates = dates.get_text().strip()
dates = dates.split("\r\n\t")
```

```python
dates = [d.replace('\t', '') for d in dates]
dates = [d.replace('\n', '') for d in dates]
```

```python
dates = list(filter(None, dates))
```

```python
dates
```

```python
link1 = "https://www.atptour.com/en/rankings/singles?rankDate="
link2 = "&rankRange=0-100"
```

```python
links = []
for i in range(1, len(dates)): # excluding dates[0] on purpose because it is a repeated date
    new_link = link1 + dates[i] + link2
    links.append(new_link)
```

```python
links
```

```python
# Excercise 1: How long has Roger Federer been in the top 5?
for l in range(0, len(links)): 
    # Outer for loop goes through the different pages. 
    # Inner for loop reads and saves data in string list.
    # If statement breaks the outer for loop if Federer isn't in the top 5 anymore.

    # redefining some vars
    link = links[l]
    response = requests.get(link, timeout = 15)
    bs=BeautifulSoup(response.content, features = "html")
    
    #reading the table
    table_body=bs.find('tbody') 
    rows = table_body.find_all('tr')
    top100 = []
    for row in rows:
        cols=row.find_all('td')
        cols=[x.text.strip() for x in cols]
        top100.append(cols)
        
    matching = [players for players in top100 if "Roger Federer" in players]
    matching = matching[0]
    
    if int(matching[0]) > 5:
        print("Since "+ dates[l])
        break
```
